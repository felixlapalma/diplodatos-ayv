```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd_w=pd.read_csv('wtageinf.csv')
pd_w.head(5)
```

```python
colLst=pd_w.columns
pLst=list(colLst[5:])
pLst
```

```python
def plt_pd_vars(x,y,xlabel,ylabel,legend,title):
    plt.plot(x,y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(legend)
    plt.title(title)
    plt.grid(True)
    plt.show()

plt_pd_vars(pd_w['Agemos'][pd_w.Sex==1],pd_w[pLst][pd_w.Sex==1],'Age [months]','Weight [Kg]',pLst,'Weight - Male')
plt_pd_vars(pd_w['Agemos'][pd_w.Sex==2],pd_w[pLst][pd_w.Sex==2],'Age [months]','Weight [Kg]',pLst,'Weight - Female')
```

## Ejercicio: Pasos Fronterizos de Argentina
Leer ubicaciones de pasos fronterizos y dibujar en un mapa.

```python
import gmplot
```

```python
pd_p=pd.read_json('pasos.json')
pd_p.head(5)
```

```python
gmap_scatter = gmplot.GoogleMapPlotter(-35.466956, -66.438481, 6)
gmap_marker = gmplot.GoogleMapPlotter(-35.466956, -66.438481, 6)
```

```python
gmap_scatter = gmplot.GoogleMapPlotter(-35.466956, -66.438481, 6)
# Esta Forma no puede tener los nombres
# https://stackoverflow.com/questions/40142959/add-text-to-scatter-point-using-python-gmplot
aux=[]
for x,y in pd_p['location']:
    aux.append((x,y))
# Scatter points
pasos_lats, pasos_lons = zip(*aux)
gmap_scatter.scatter(pasos_lats, pasos_lons, '#3B0B39', size=16000, marker=False)
gmap_scatter.draw("pasos_sin_marcadores.html")
```

```python
gmap_marker = gmplot.GoogleMapPlotter(-35.466956, -66.438481, 6)
# Esta Forma SI puede tener los nombres
# https://stackoverflow.com/questions/40142959/add-text-to-scatter-point-using-python-gmplot
aux=[]
titles=list(pd_p['name']);
j=-1
for x,y in pd_p['location']:
    j=j+1
    # Scatter points
    gmap_marker.marker(x, y,title=titles[j])
gmap_marker.draw("pasos_con_marcadores.html")
```

## Ejercicio: Representación de grafos

```python
import numpy as np
edgeList = [ [0, 2], [1, 3], [2, 3],[2, 4], [3, 5], [4, 5] ];
```

```python
flat_list = list(set([item for sublist in edgeList for item in sublist]))
flat_list
```

### Matriz de adyacencia

La matriz de adyacencia es una matriz cuadrada que se utiliza como una forma de representar relaciones binarias.- https://es.wikipedia.org/wiki/Matriz_de_adyacencia

```python
# Init array
ady_mat=np.zeros((len(flat_list), len(flat_list)),dtype=int)
ady_mat
```

```python
for k in edgeList:
    ady_mat[k[0],k[1]]=1
ady_mat
```

### Lista de adyacencia
*https://es.wikipedia.org/wiki/Lista_de_adyacencia (https://www.boost.org/doc/libs/1_43_0/libs/graph/doc/adjacency_list.html) --En teoría de grafos, una lista de adyacencia es una representación de todas las aristas o arcos de un grafo mediante una lista.
Si el grafo es no dirigido, cada entrada es un conjunto o multiconjunto de dos vértices conteniendo los dos extremos de la arista correspondiente. Si el grafo es dirigido, cada entrada es una tupla de dos nodos, uno denotando el nodo fuente y el otro denotando el nodo destino del arco correspondiente. Típicamente, las listas de adyacentes no son ordenadas.*

```python
# Dado que es un grafo dirigido, aprovechamos la matriz de adyacencia
ady_list=[]
flat_list_np=np.array(flat_list)
for j in range(0,ady_mat.shape[0]):
    ady_list.append(flat_list_np[list(ady_mat[j]>0)])
ady_list
```
